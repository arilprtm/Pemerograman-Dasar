```python
import matplotlib.pyplot as plt

bakery = ('Manique', 'Sweetly', 'Doughla', 'Cakely', 'Sugary',
        'Brielle','Crav√©', 'Delice', 'Bakea')

customer = (45, 11, 21, 19, 50, 17,
            34, 28, 78)

plt.bar(x=bakery, height=customer)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.barh(y=bakery, width=customer)
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({
    'Bakery': bakery,
    'Customer': customer,
})

df.sort_values(by='Customer', inplace=True)

plt.barh(y=df["Bakery"], width=df["Customer"])
plt.show()
```

```python
plt.barh(y=df["Bakery"], width=df["Customer"])
plt.xlabel("Customer (Orang)")
plt.title("Jumlah Customer Bakery")
plt.show()
```

```python
import seaborn as sns

sns.barplot(y=df["Bakery"], x=df["Customer"], orient="h", color='blue')
plt.barh(y=df["Bakery"], width=df["Customer"])
plt.xlabel("Customer (Orang)")
plt.title("Jumlah Customer Bakery")
plt.show()
```

```python
import matplotlib.pyplot as plt

cakes = ('Cheesecake', 'Cookies', 'Red Velvet', 'Others')
votes = (50, 20, 30, 10)
colors = ('#FFF8DC', '#8B4513', '#a91137', '#E67F0D')
explode = (0, 0, 0, 0)

plt.pie(
    x=votes,
    labels=cakes,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode
)
plt.show()
```

```python
plt.pie(
    x=votes,
    labels=cakes,
    colors=colors,
    wedgeprops = {'width': 0.5}
)
plt.show()
```

```python
import matplotlib.pyplot as plt

lemon_diameter = [6.44, 6.87, 7.7, 8.85, 8.15,
                  9.96, 7.21, 10.04, 10.2, 11.06]
lemon_weight = [112.05, 114.58, 116.71, 117.4, 128.93,
                132.93, 138.92, 145.98, 148.44, 152.81]

plt.scatter(lemon_diameter, lemon_weight)
plt.xlabel("Diameter Lemon")
plt.ylabel("Berat Lemon")
plt.title("Hubungan Diameter dan Berat Lemon")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x=lemon_diameter, y=lemon_weight)
plt.xlabel("Diameter Lemon")
plt.ylabel("Berat Lemon")
plt.title("Hubungan Diameter dan Berat Lemon")
plt.show()
```

```python
sns.regplot(x=lemon_diameter, y=lemon_weight)
plt.xlabel("Diameter Lemon")
plt.ylabel("Berat Lemon")
plt.title("Hubungan Diameter dan Berat Lemon")
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Unduh data historis saham BBCA.JK
ticker = yf.Ticker("BBCA.JK")
df = ticker.history(period="6mo", interval="1d")  # atau bisa diubah ke "1y", dll

# Reset index agar tanggal menjadi kolom
df.reset_index(inplace=True)

# Visualisasi harga penutupan
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Close'], color='red')
plt.xlabel('Date', fontsize=15)
plt.ylabel('Close Price (IDR)', fontsize=15)
plt.title('Harga Saham BBCA.JK', fontsize=16)
plt.grid(True)
plt.tight_layout()
plt.show()
```
