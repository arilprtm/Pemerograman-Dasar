```python
import pandas as pd
```

```python
df = pd.read_csv('/content/orders.csv')
```

```python
# Ringkasan statistik
df.describe()

# Rata-rata pembayaran
rata_rata_pembayaran = df['payment'].mean()
print("Rata-rata pembayaran:", rata_rata_pembayaran)

# Modus jumlah
modus_customer_id = df['customer_id'].mode()
print("Modus customer_id:", modus_customer_id)
```

```python
import matplotlib.pyplot as plt

df['delivery_date'].hist()
plt.title("Distribusi Durasi Pengiriman (Hari)")
plt.xlabel("Hari")
plt.ylabel("Frekuensi")
plt.show()
```

```python
# Mean
group_mean_customer_head = df.groupby("customer_id").mean(numeric_only=True)
group_mean_customer_head.sort_values(by='payment', ascending=False).head(5)

# Agg lengkap: Menghitung jumlah customer per kelompok pembayaran, mengurutkan berdasarkan jumlah customer terbanyak, dan menampilkan 5 teratas.
group_agg_customer = df.groupby("customer_id").agg({
    'payment': ['count']
})
group_agg_customer.sort_values(('payment', 'count'), ascending=False).head(5)
```

```python
# Korelasi
corr_matrix = df[['payment', 'customer_id']].corr()
print("Korelasi:", corr_matrix)

# Kovarian
cov_matrix = df[['payment', 'customer_id']].cov()
print("Kovarian:", cov_matrix)
```
